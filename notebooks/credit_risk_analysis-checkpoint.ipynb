## Pré-processamento de Dados.

```python
# Conversão de variáveis categóricas.
df["Estado Civil"] = df['Estado Civil'].map({'Solteiro': 0, 'Casado': 1, 'Divorciado': 2})
df['Histórico de Crédito'] = df['Histórico de Crédito'].map({'Negativo': 0, 'Positivo': 1, 'Sem Histórico': 2})
```

```python
# Número de colunas no seu dataframe
num_columns = len(df.columns)

# Definindo a quantidade de linhas e colunas para a grid de subgráficos
ncols = 3  # Número de colunas
nrows = np.ceil(num_columns / ncols).astype(int)  # Número de linhas necessário

# Plotando gráficos de distribuição (histogramas) para cada variável
plt.figure(figsize=(ncols*5, nrows*5))  # Ajustando o tamanho da figura com base no número de gráficos
for i, col in enumerate(df.columns, 1):
    plt.subplot(nrows, ncols, i)  # Ajustando o layout dinamicamente
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribuição de {col} Antes da Normalização')

plt.tight_layout()
plt.show()
```

```python
# Normalização Min-Max.
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[['Renda Anual', 'Empréstimo Atual', 'Número de Dependentes', 
                                                     'Número de Cartões de Crédito', 'Valor de Dívida Total', 
                                                     'Score de Crédito']]), 
                            columns=['Renda Anual', 'Empréstimo Atual', 'Número de Dependentes', 
                                     'Número de Cartões de Crédito', 'Valor de Dívida Total', 
                                     'Score de Crédito'])
```

```python
# Padronização Z-Score: A padronização transforma os dados para que tenham média 0 e desvio padrão 1.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df[['Renda Anual', 'Empréstimo Atual', 'Número de Dependentes', 
                                                        'Número de Cartões de Crédito', 'Valor de Dívida Total', 
                                                        'Score de Crédito']]), 
                               columns=['Renda Anual', 'Empréstimo Atual', 'Número de Dependentes', 
                                        'Número de Cartões de Crédito', 'Valor de Dívida Total', 
                                        'Score de Crédito'])
```

```python
# Verificar os valores mínimo e máximo das variáveis normalizadas
print(df_normalized.min())  # Todos devem ser 0 ou próximos de 0
print(df_normalized.max())  # Todos devem ser 1 ou próximos de 1
```

```python
# Verificação após Padronização Z-Score (Média = 0, Desvio Padrão = 1).
# Verificar a média e o desvio padrão das variáveis padronizadas
print(df_standardized.mean())  # Todos devem ser 0 ou próximos de 0
print(df_standardized.std())   # Todos devem ser 1 ou próximos de 1
```

```python
# Verificar a média e o desvio padrão das variáveis padronizadas
print(df_standardized.mean())  # Todos devem ser 0 ou próximos de 0
print(df_standardized.std())   # Todos devem ser 1 ou próximos de 1
```
